## Statistical Analysis of the Impact of Nitrous Oxide Emissions on Agricultural Yield

```python
df = pd.read_csv('data.csv')
```

```python
plt.figure(figsize = (10,10))
plt.subplot(211)
plt.title('Average of Yield (t/ha) by year')
plt.plot(df['year'], df['Average of Yield (t/ha)'],'go-')
plt.subplot(212)
plt.title('Sum of NitrousEmission by year')
plt.plot(df['year'], df['Sum of NitrousEmission'],'ro-' )
plt.savefig('yieldEmissionTrend');
```

## Hypotesis Testing



__Null Hypothesis (H0)__: There is no effect of nitrous oxide emissions on agricultural yield. 

__Alternative Hypothesis (H1)__: There is an effect of nitrous oxide emissions on agricultural yield.

If the p-value < 0.05, reject the null hypothesis and conclude there is a significant relationship.
If the p-value ≥ 0.05, do not reject the null hypothesis, indicating insufficient evidence of a significant relationship

```python
slope, intercept, r_value , p_value, std_err = linregress(df['Sum of NitrousEmission'],df['Average of Yield (t/ha)'])
```

```python
# Hypothesis Testing Results
hypothesis_test_results = {
    'Slope': slope,
    'Intercept': intercept,
    'R-squared': r_value**2,
    'P-value': p_value,
    'Standard Error': std_err
}
```

```python
print("Hypothesis Testing Results:")
for key, value in hypothesis_test_results.items():
    print(f"{key}: {value}")
```

__Check for correlation__

```python
df.iloc[:,1:].corr()
```

```python
import scipy.stats as stats
correlation, p_value = stats.pearsonr(df['Sum of NitrousEmission'],df['Average of Yield (t/ha)'])
```

```python
sns.regplot(y=df['Average of Yield (t/ha)'],x=df['Sum of NitrousEmission'])
plt.xlabel('Nitrous Oxide Emissions')
plt.ylabel('Agricultural Yield')
plt.title(f'R-Squared: {r_value**2:.2f}, P-value: {p_value:.2f}, correlation:{correlation:.2f}')
plt.savefig('NitrousVSAgricyield.png')
```


**Hypothesis Testing Results:**

**Slope: 2.898**

This indicates that for every metric ton of CO2 equivalent of NO2 emmited, the agricultural yield changes by an  approximately 2.897  tonnes per hectare on average  in the same year.

**Intercept: -3.626 **

This is the expected change in agricultural yield when the change in nitrous oxide emissions is zero.

**R-squared:  0.96**

This value suggests that approximately 96% of the variance in agricultural yield changes can be explained by the changes in nitrous oxide emissions.

**P-value:  5.57**

This p-value is less than 0.05, indicating that the relationship between nitrous oxide emissions and agricultural yield changes is statistically significant. Therefore, we reject the null hypothesis that there is no relationship between nitrous oxide emissions and agricultural yield.


## Summary

The linear regression analysis shows that there is statistically significant relationship between nitrous oxide emissions and agricultural yield changes, which indicate that nitrous oxide emissions impact agricultural yield. The positive slope suggests that increases in nitrous oxide emissions are associated with increases in agricultural yield in the same year also the R-squared value also indicates that 96% of agricultural yield changes can be explained by nitrous emission.
